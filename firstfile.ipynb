# Exploratory Data Analysis

### Gender

```python
#count plot for Gender
ax = sns.countplot(x='Gender',data=df)
```

```python
#count plot for Gender with total values of females and males 
ax = sns.countplot(x='Gender',data=df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group kar rhe male and female ko aur toatal amount nikal rhe sum function ko use kar ke ek gender ke kitna total spend kiya
df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
#Bar plot for amount of female and male
sales_gender =df.groupby(['Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x='Gender',y='Amount', data= sales_gender)
```

*From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men*

## AGE

```python
#Count graph plot kar rhe Age Group ka separated by Gender 
ax = sns.countplot(data=df, x='Age Group', hue='Gender')
```

```python
#count plot for Age Group separated by Gender with total values of females and males 
ax = sns.countplot(data=df, x='Age Group', hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group kar rhe Age Group ne kitna toatal amount spend kiya
#sum function ko use kar ke ek Age Gender ne kitna total spend kiya
df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

```python
#Bar plot for Age Group vs Total Spend
sales_age =df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.barplot(x='Age Group',y='Amount', data= sales_age)
```

*From above graphs we can see that most of the buyers are of age group between 26-35 yrs female*

## State

```python
# total number of orders from top 10 states
sales_state =df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
#Graph ka size set kar rhe kyuke default size me name overlap ho rhe
sns.set(rc={'figure.figsize':(15,5)})
#Bar graph plot ko kar rhe hai 
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

```python
# total amount/sales spend by top 10 states
sales_state =df.groupby(['State'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
#Graph ka size set kar rhe kyuke default size me name overlap ho rhe
sns.set(rc={'figure.figsize':(15,5)})
#Bar graph plot ko kar rhe hai 
sns.barplot(data = sales_state, x = 'State',y= 'Amount')
```

*From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


## Martial Status

```python
df.columns
```

```python
ax= sns.countplot(data=df, x='Shadi Hogye')
sns.set(rc={'figure.figsize':(7,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group kar rhe hai Shadi Hogye hai ke nhi with respect to Gender and total amount ko sum up kar rhe 
df.groupby(['Shadi Hogye','Gender'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
```

*From above graphs we can see that most of the buyers are married (women) and they have high purchasing power*

## Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
occupation_order = df['Occupation'].value_counts().index
ax = sns.countplot(data=df, x='Occupation', order= occupation_order)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group kar rhe hai Occupation ko and total amount ko sum up kar rhe 
sales_Occupation =df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
#Graph ka size set kar rhe kyuke default size me name overlap ho rhe
sns.set(rc={'figure.figsize':(20,5)})
#Bar graph plot ko kar rhe hai 
sns.barplot(data = sales_Occupation, x = 'Occupation',y= 'Amount')
```

*From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector*

## Product Category

```python
sns.set(rc={'figure.figsize':(25,8)})
occupation_order = df['Product_Category'].value_counts().index
ax = sns.countplot(data=df, x='Product_Category', order= occupation_order)


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Group kar rhe hai Product Category ko and total amount ko sum up kar rhe 
sales_Product_Category =df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
#Graph ka size set kar rhe kyuke default size me name overlap ho rhe
sns.set(rc={'figure.figsize':(25,8)})
#Bar graph plot ko kar rhe hai 
sns.barplot(data = sales_Product_Category, x = 'Product_Category',y= 'Amount')
```

*From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category*

## Product ID

```python
#Top selling product ko product id se dekhenge
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```

```python
# top 10 most sold products (same thing as above)
#bss same cheez kar ke different tareka hai 
fig1, ax1 = plt.subplots(figsize=(12,7))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
```
